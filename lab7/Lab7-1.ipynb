```python
def คำนวณเงินเดือน(ชื่อ, เงินเดือน, ชั่วโมงทำงาน):
    อัตราค่าชั่วโมง = 100
    if ชั่วโมงทำงาน > 40:
        ค่าล่วงเวลา = (ชั่วโมงทำงาน - 40) * อัตราค่าชั่วโมง * 1.5 + 40 * อัตราค่าชั่วโมง
    else:
        ค่าล่วงเวลา = ชั่วโมงทำงาน * อัตราค่าชั่วโมง
    เงินเดือนรวม = เงินเดือน + ค่าล่วงเวลา
    return ชื่อ, เงินเดือน, ชั่วโมงทำงาน, ค่าล่วงเวลา, เงินเดือนรวม
ชื่อพนักงาน = input("กรุณาใส่ชื่อพนักงาน: ")
เงินเดือน = float(input("กรุณาใส่เงินเดือน (บาท): "))
ชั่วโมงทำงาน = float(input("กรุณาใส่ชั่วโมงทำงานสำหรับค่าล่วงเวลา: "))
ชื่อ, เงินเดือน, ชั่วโมงทำงาน, ค่าล่วงเวลา, เงินเดือนรวม = คำนวณเงินเดือน(ชื่อพนักงาน, เงินเดือน, ชั่วโมงทำงาน)
print("---------------------------------")
print(f"ชื่อพนักงาน: {ชื่อ}")
print(f"เงินเดือน: {เงินเดือน:.2f} บาท")
print(f"ชั่วโมงล่วงเวลา: {ชั่วโมงทำงาน:.0f} ชั่วโมง")
print(f"ค่าล่วงเวลา: {ค่าล่วงเวลา:.2f} บาท")
print(f"รวมเป็นเงินเดือนทั้งหมด: {เงินเดือนรวม:.2f} บาท")
```
